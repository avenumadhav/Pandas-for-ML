### Q4. Selection
1. Retrieve part of the dataframe based on condition: df['Country'] == 'Japan'
2. Retrieve part of the dataframe based on conditions: df['Country'] == 'Japan' and df['City proper[c] Area (km2)'] > 200
3. Retrieve part of the dataframe based on conditions: New city Indian municipalities with UN 2018 population > 10000000

```python
df.columns
```

```python
#1
filt = df['Country'] == 'Japan'
df[filt]
```

```python
#2
filt1 = df['Country'] == 'Japan' 
filt2 = df['City proper[c] Area (km2)'] > 200

filt = filt1 & filt2
df[filt]
```

```python
#3
filt1 = df['Country'] == 'India' 
filt2 = df['UN 2018 population estimates[b]'] > 10000000
filt3 = df['City proper[c] Definition'] == 'Municipality'

filt = filt1 & filt2 & filt3
df[filt]
```
